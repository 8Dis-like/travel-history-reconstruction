```python
import os
import yaml

# Update this path if your Securiport folder is located elsewhere in your Drive
project_dir = '/content/drive/MyDrive/Securiport'
os.chdir(project_dir)

yaml_path = 'dataset.yaml'
with open(yaml_path, 'r') as f:
    config = yaml.safe_load(f)

config['path'] = f"{project_dir}/data/synthetic_dataset"

with open(yaml_path, 'w') as f:
    yaml.dump(config, f, default_flow_style=False)

print(f"Working directory set to: {os.getcwd()}")
print("dataset.yaml paths updated successfully!")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')

results = model.train(
    data='dataset.yaml',
    epochs=100,
    imgsz=640,
    batch=64,
    device='0',
    workers=8,
    project='runs/detect',
    name='a100_finetune'
)
```
